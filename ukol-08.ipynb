### Bonus

Adopce kterých zvířat je nejdražší?

```python
zvirata_podle_ceny = zvirata.sort_values("cena", ascending=False)
nejvyssi_cena = zvirata_podle_ceny.iloc[0]["cena"]
nejdrazsi_zvirata = zvirata[zvirata["cena"]==nejvyssi_cena]
nejdrazsi_zvirata[["nazev_cz", "nazev_en"]]
```

Které zvíře je alfabeticky poslední v češtině? 

```python
zvirata_podle_abc_cz = zvirata.sort_values("nazev_cz", ascending=False)
zvirata_podle_abc_cz.iloc[0]
```

Které v angličtině?

```python
zvirata_podle_abc_en = zvirata.sort_values("nazev_en", ascending=False)
zvirata_podle_abc_en.iloc[0]
```

Zvířata které třídy jsou v průměru nejlevnější a nejdražší?

```python
tridy_prumerna_cena = zvirata.groupby("trida_cz")["cena"].mean()
tridy_prumerna_cena.sort_values()
```

Která zvířata nejsou k dispozici k prohlídce?

```python
neni_prohlidka = zvirata[zvirata["k_prohlidce"]==0]
neni_prohlidka
```

Kolik zvířat není k dispozici k prohlídce?

```python
neni_prohlidka.shape[0]
```

Kolik zvířat které třídy není k dispozici k prohlídce?

```python
neni_prohlidka.groupby("trida_cz").count()
```

Ve které třídě je nejvíc a ve které nejmíň zvířat?

```python
tridy = zvirata.groupby("trida_cz").count()
tridy.sort_values("id")
```
